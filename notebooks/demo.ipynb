```julia
using LinearAlgebra
using Plots, LaTeXStrings

using Gershgorin
```

```julia
# Make a random (5,5) real matrix
M = randn(ComplexF32, 5, 5)

gershgorin(M; c=:blue, label=L"$M$")  # Plot Gershgorin's discs
 = eigvals(M)
p1 = plot!(, seriestype=[:scatter], c=:black, label=L"$\lambda(M)$")

gershgorin(transpose(M); c=:red, label=L"$M^T$")  # Now do the same for the transpose
 = eigvals(transpose(M))
p2 = plot!(, seriestype=[:scatter], c=:black, label=L"$\lambda(M^T)$")

plot(p1, p2, link=:all) |> display
savefig("demo.png")
```
