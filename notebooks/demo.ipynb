```julia
using LinearAlgebra
using Plots, LaTeXStrings

using Gershgorin
```

```julia
# Make a random (5,5) real matrix
M = randn(ComplexF32, 5, 5)

gershgorin(M; c=:blue, label=L"$G(M)$")  # Plot Gershgorin's discs
λ = eigvals(M)
p1 = plot!(λ, seriestype=[:scatter], c=:black, label=L"$\lambda(M)$")

gershgorin(transpose(M); c=:red, label=L"$G(M^T)$")  # Now do the same for the transpose
λ = eigvals(transpose(M))
p2 = plot!(λ, seriestype=[:scatter], c=:black, label=L"$\lambda(M^T)$")

gershgorin(M; c=:blue)
gershgorin!(transpose(M); c=:red)
overlap(M, transpose(M), c=:black, alpha=1, label=L"G(M) \cap G(M^T)")
p3 = plot!(λ, seriestype=[:scatter], c=:white, label=L"$\lambda(M)$")

plot(p1, p2, p3, link=:all, dpi=300, layout=(1,3), size=(750,350)) |> display
savefig("demo.png")
```
