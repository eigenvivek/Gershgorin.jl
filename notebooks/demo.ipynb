```julia
using LinearAlgebra
using Plots, LaTeXStrings

using Gershgorin
```

```julia
# Make a random (5,5) real matrix
M = randn(5, 5)

gershgorin(M; c=:blue, label=L"$M$") 
λ = eigvals(M)
p1 = plot!(λ, seriestype=[:scatter], c=:black, label=L"$\lambda(M)$")

gershgorin(M'; c=:red, label=L"$M'$")
λ = eigvals(M')
p2 = plot!(λ, seriestype=[:scatter], c=:black, label=L"$\lambda(M')$")

plot(p1, p2, link=:all) |> display
savefig("demo.png")
```
